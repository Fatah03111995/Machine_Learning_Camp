## MISSING VALUE

```python
# kolom yang memiliki baris not null paling banyak, memiliki jumlah baris 1460
bounderies = 1460 * 0.75
print('batasan 75%, sebesar', bounderies)
```

```python
less = missing_values[missing_values<1095].index
over = missing_values[missing_values>= 1095].index
# .index: Bagian ini mengambil indeks dari elemen-elemen yang sudah difilter di langkah pertama.
# Indeks ini biasanya merupakan nama kolom dari DataFrame missing_values.
```

## untuk data yang hilang < 75%
## mengisi nilai yang hilang dengan nilai median

```python
numeric_feature=train_df[less].select_dtypes(include=['number']).columns
# metode select_dtypes() digunakan untuk memilih kolom yang memiliki tipe data numerik (seperti int64, float64).
# .columns: Bagian ini mengambil nama kolom dari subset DataFrame yang sudah difilter tadi,
# menghasilkan daftar nama kolom yang bertipe numerik.
```

```python
train_df[numeric_feature] = train_df[numeric_feature].fillna(train_df[numeric_feature].median())
# mengisi data kosong di kolom yang termasuk numeric feature dengan data median
# Pada kode ini:
# train_df[numeric_feature].median() menghitung median dari setiap kolom numerik di numeric_feature.
# train_df[numeric_feature].fillna(...) mengisi NaN di semua kolom numerik dengan nilai median masing-masing kolom, sekaligus.
# Potensi Masalah: Jika train_df[numeric_feature] berisi beberapa kolom dan ada kolom dengan tipe data campuran,
# maka fillna() dalam satu operasi ini mungkin tidak berhasil karena bisa terjadi masalah dalam casting tipe data.
# Tetapi jika semua kolom di numeric_feature adalah numerik murni, ini akan berjalan lancar.
```

```python
# lebih aman menggunakan
# for column in numerical_feature:
#     train_df[column] = train_df[column].fillna(train_df[column].median())
```

```python
kategorical_feature = train_df[less].select_dtypes(include='object').columns
# ERROR - train_df[kategorical_feature] = train_df[kategorical_feature].fillna(train_df[kategorical_feature].mode()[0])
for column in kategorical_feature :
    train_df[column] = train_df[column].fillna(train_df[column].mode()[0])
```

```python
# NOTES
# Pemilihan metode pengisian nilai yang hilang (NaN) menggunakan mean, median, atau mode bergantung pada jenis data dan
# distribusi data dalam kolom tersebut. Berikut penjelasan mengenai kapan sebaiknya menggunakan masing-masing metode:

# Menggunakan Mean (Rata-Rata):
# Kapan digunakan: Gunakan rata-rata (mean) untuk mengisi nilai hilang jika data memiliki distribusi yang simetris atau
# mendekati distribusi normal (misalnya, data tinggi badan atau berat badan dalam populasi besar).
# Keuntungan: Memberikan perkiraan nilai yang paling representatif untuk data yang simetris dan tidak memiliki pencilan (outliers).
# Kapan dihindari: Hindari menggunakan mean jika data memiliki pencilan yang signifikan atau distribusi yang sangat miring,
# karena pencilan dapat menyebabkan rata-rata menjadi kurang representatif dan memengaruhi hasil.


# Menggunakan Median:
# Kapan digunakan: Gunakan median ketika data memiliki distribusi yang miring atau terdapat banyak pencilan
# (misalnya, data pendapatan yang sering kali memiliki distribusi miring).
# Keuntungan: Median lebih tahan terhadap pencilan dan distribusi data yang tidak normal, sehingga lebih representatif untuk data yang tidak simetris.
# Contoh penggunaan: Untuk kolom seperti harga properti atau pendapatan yang memiliki pencilan atau distribusi yang miring.


# Menggunakan Mode:
# Kapan digunakan: Gunakan mode (nilai yang paling sering muncul) untuk data kategorikal atau
# kolom yang memiliki sedikit variasi nilai unik (misalnya, kolom yang hanya memiliki beberapa kategori seperti "Jenis Kelamin",
# "Status Pernikahan", atau "Kategori Produk").
# Keuntungan: Mode memungkinkan pengisian nilai hilang berdasarkan nilai yang paling umum, yang masuk akal untuk data kategorikal dan nominal.
# Kapan dihindari: Mode kurang cocok untuk data numerik yang memiliki banyak nilai unik atau jika kolom tersebut seharusnya memiliki
# nilai yang berbeda-beda secara signifikan.

# Ringkasan
# Mean: Gunakan untuk data numerik dengan distribusi simetris dan tanpa pencilan.
# Median: Gunakan untuk data numerik dengan distribusi yang miring atau banyak pencilan.
# Mode: Gunakan untuk data kategorikal atau data dengan sedikit variasi nilai.
# Memilih metode yang tepat akan menghasilkan data yang lebih akurat dan representatif setelah pengisian nilai yang hilang.
```

```python
condition = ~((new_df[numeric_newdf] < (q1 - 1.5*iqr)) | (new_df[numeric_newdf] > (q3 + 1.5 * iqr))).any(axis=1)
new_df_filterednumeric = new_df.loc[condition, numeric_newdf]

for feature in numeric_newdf:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=new_df_filterednumeric[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

### label encoding

```python
from sklearn.preprocessing import LabelEncoder
import math

label_encoder = LabelEncoder()
df_lencoder = pd.DataFrame(new_df)

for col in category_df:
    df_lencoder[col] = label_encoder.fit_transform(new_df[col])

df_lencoder
```

```python
# menentukan jumlah baris dan kolom untuk grid subplot
num_vars = df_lencoder.shape[1] #jml variable
n_cols = 4 # jml kolom yang diinginkan
n_rows = math.ceil(num_vars / n_cols) # jml baris
n_rowa = -(-num_vars // n_cols) #ceiling tanpa math.ceil
print(n_rowa)

#membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows*4))

# flatten axes array untuk memudahkan iterasi jika diperlukan
axes = axes.flatten()

# plot setiap variabel
for i, column in enumerate(df_lencoder.columns):
    df_lencoder[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('value')
    axes[i].set_ylabel('frequency')

# menghapus subplot yang tidak terpakai jika ada
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```
